```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import f1_score, confusion_matrix
import pandas as pd
all_data = loadmat('data/heartbeat_data/000.mat')
all_data.keys()
all_data['val']
# Access the 'val' key
val_data = all_data['val']
```

```python
print("Dimensions of val_data:", val_data.shape)
```

```python
# Get rid of the first dimension since it has only one element
val_data = val_data[0]

# Plot each channel separately
for i in range(val_data.shape[0]):
    plt.figure()
    plt.plot(val_data[i])
    plt.title(f'Channel {i+1}')
    plt.xlabel('Sample Index')
    plt.ylabel('Value')
    plt.show()
```
